```python
# =======================
# 1. IMPORT
from google.colab import files
uploaded = files.upload()
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
from scipy.cluster.hierarchy import dendrogram, linkage

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA

# =======================

# =======================
# 2. LOAD DATA
# =======================
df = pd.read_csv('churn-bigml-80.csv')

# =======================
# 3. CLEAN DATA
# =======================
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

df.rename(columns={'churn': 'churned'}, inplace=True)
df.rename(columns={
    'international_plan': 'intl_plan',
    'customer_service_calls': 'number_customer_service_calls'
}, inplace=True)

df.dropna(inplace=True)

# drop cột không cần
df.drop(['state','area_code','phone_number'], axis=1, inplace=True, errors='ignore')
# =======================
# 4. ENCODE DATA
# =======================
for col in df.select_dtypes(include=['object', 'bool']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# =======================
# 6. SPLIT FEATURES/TARGET
# =======================
X_train = df.drop('churned', axis=1)
y_train = df['churned']


print("X_train:", X_train.shape)
print("y_train:", y_train.shape)

# =======================
# 7. SCALE DATA
# =======================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

print("Scaled data shape:", X_train_scaled.shape)

# =======================
# 8. PCA FOR VISUALIZATION
# =======================
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)

# =======================
# 9. K-MEANS
# =======================
print("\nRunning K-Means...")

inertia = []
sil_scores = []
K_range_inertia = range(1, 10)
K_range_sil = range(2, 10)

for k in K_range_inertia:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_train_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(7, 4))
plt.plot(K_range_inertia, inertia, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia (WCSS)")
plt.title("Elbow Method")
plt.tight_layout()
plt.show()

for k in K_range_sil:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_k = km.fit_predict(X_train_scaled)
    sil_scores.append(silhouette_score(X_train_scaled, labels_k))

plt.figure(figsize=(7, 4))
plt.plot(K_range_sil, sil_scores, marker='o')
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Method")
plt.tight_layout()
plt.show()

k_opt = 3
kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init=10)
clusters_kmeans = kmeans.fit_predict(X_train_scaled)

inertia_final = kmeans.inertia_
silhouette_final = silhouette_score(X_train_scaled, clusters_kmeans)
dbi_final = davies_bouldin_score(X_train_scaled, clusters_kmeans)

print("\n===== K-Means Evaluation =====")
print("Inertia:", inertia_final)
print("Silhouette Score:", silhouette_final)
print("Davies-Bouldin Index:", dbi_final)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_kmeans, alpha=0.7)

centroids = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='X')

plt.title("K-Means Clustering (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

df['cluster_kmeans'] = clusters_kmeans
print("\nK-Means cluster counts:")
print(df['cluster_kmeans'].value_counts())
print("\nK-Means cluster means:")
print(df.groupby('cluster_kmeans').mean(numeric_only=True))

# =======================
# 10. HIERARCHICAL CLUSTERING
# =======================
print("\nRunning Hierarchical Clustering...")

hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters_hc = hc.fit_predict(X_train_scaled)

print("Hierarchical clusters:", np.unique(clusters_hc))

hc_sil = silhouette_score(X_train_scaled, clusters_hc)
hc_dbi = davies_bouldin_score(X_train_scaled, clusters_hc)

print("\n===== Hierarchical Evaluation =====")
print("Silhouette Score:", hc_sil)
print("Davies-Bouldin Index:", hc_dbi)

linked = linkage(X_train_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title("Dendrogram - Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_hc, alpha=0.7)
plt.title("Hierarchical Clustering (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

df['cluster_hierarchical'] = clusters_hc
print("\nHierarchical cluster counts:")
print(df['cluster_hierarchical'].value_counts())

# =======================
# 11. DBSCAN
# =======================
print("\nRunning DBSCAN tuning...")

for eps in [0.5, 1.0, 1.5, 2.0]:
    dbscan_tmp = DBSCAN(eps=eps, min_samples=5)
    labels_tmp = dbscan_tmp.fit_predict(X_train_scaled)

    n_clusters = len(set(labels_tmp)) - (1 if -1 in labels_tmp else 0)
    n_noise = list(labels_tmp).count(-1)

    print(f"eps = {eps}")
    print("Clusters:", np.unique(labels_tmp))
    print("Number of clusters:", n_clusters)
    print("Noise points:", n_noise)
    print("-" * 30)

dbscan = DBSCAN(eps=1.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(X_train_scaled)

n_clusters = len(set(clusters_dbscan)) - (1 if -1 in clusters_dbscan else 0)
n_noise = list(clusters_dbscan).count(-1)

print("\nFinal DBSCAN clusters:", np.unique(clusters_dbscan))
print("Number of clusters:", n_clusters)
print("Noise points:", n_noise)

plt.figure(figsize=(8, 6))
unique_labels = np.unique(clusters_dbscan)

for label in unique_labels:
    mask = clusters_dbscan == label
    if label == -1:
        plt.scatter(X_pca[mask, 0], X_pca[mask, 1], c='red', label='Noise', alpha=0.6)
    else:
        plt.scatter(X_pca[mask, 0], X_pca[mask, 1], label=f'Cluster {label}', alpha=0.6)

core_samples = dbscan.core_sample_indices_
plt.scatter(X_pca[core_samples, 0], X_pca[core_samples, 1],
            s=150, facecolors='none', edgecolors='black',
            linewidths=1.5, label='Core points')

plt.title(f"DBSCAN (eps=1.5)\nClusters={n_clusters}, Noise={n_noise}")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.tight_layout()
plt.show()

eps_values = [0.5, 1.0, 1.5, 2.0]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, eps in zip(axes.flatten(), eps_values):
    dbscan = DBSCAN(eps=eps, min_samples=5)
    labels = dbscan.fit_predict(X_train_scaled)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)

    ax.scatter(X_pca[:,0], X_pca[:,1], c=labels, alpha=0.6)
    ax.set_title(f"eps = {eps}")


plt.tight_layout()
plt.show()


df['cluster_dbscan'] = clusters_dbscan
print("\nDBSCAN cluster counts:")
print(df['cluster_dbscan'].value_counts())

print("\nClustering done.")
```
